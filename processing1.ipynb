## Date Processing

```python
all_date = pd.DataFrame({'all_date':pd.date_range('1/1/2023','31/12/2023')})
all_date.head()
```

```python
#dayofweek: 0-Mon ---> 6-Sun
weekend_cond_ = (all_date['all_date'] .dt .dayofweek != 5) \
                & (all_date['all_date'] .dt .dayofweek != 6)
all_date = all_date .loc[weekend_cond_, :] .copy()
all_date
```

```python
all_date.head(10)
```

```python
all_date['Month'] = all_date['all_date'].dt.month
```

```python
# .isocalendar().week = tuan 1 den tuan 52
all_date['Week'] = all_date['all_date'].dt.isocalendar().week
```

```python
all_date.head(20)
```

```python
ref_date = pd.DataFrame()
for _ in range(1 , 13):
    cond_ = all_date['Month'] == _
    
    each_month = all_date.loc[cond_, :].copy()
    each_month['Week number'] = each_month['Week'] - (each_month['Week'].min() - 1)
    
    ref_date = pd.concat([ref_date, each_month])
```

```python
ref_date.tail(30)
```

### Details by Week

```python
pj_by_week = pd.merge(allc, ref_date
                      ,how = 'inner'
                      ,left_on = ['project_duration']
                      ,right_on = ['all_date'])
```

```python
pj_by_week.head(10)
```

```python
def get_month():
    month = input('Enter a month (number): ')
    while True:
        if month.isdigit():
            month = int(month)
            if (month >= 1) & (month <= 12):
                break
            else:
                month = input('Invalid month number, please enter again: ')
        else:
            month = input('Invalid month number, please enter again: ')
    return int(month)

def get_week():
    week = input('Enter a week (number): ')
    while True:
        if week.isdigit():
            week = int(week)
            if (week >= 1) & (week <= 6):
                break
            else:
                week = input('Invalid week number, please enter again: ')
        else:
            week = input('Invalid week number, please enter again: ')
    return int(week)

```

```python
MONTH = get_month()
WEEK_NUMBER = get_week()
```

```python
get_list_row_cond_ = (pj_by_week['Month'] == MONTH)\
                    & (pj_by_week['Week number'] == WEEK_NUMBER)

get_list_by_week = pj_by_week.loc[get_list_row_cond_, :].copy()
get_list_by_week.head()
```

```python
get_list_by_week.shape
```

```python
allocated_acc_cond = (allocated.loc[(allocated['Month'] == MONTH)\
                                   & (allocated['Week number'] == WEEK_NUMBER)
                                   , ['Acc']])['Acc'].unique()

not_full_allocated_acc_cond = (not_f_allocated.loc[(not_f_allocated['Month'] == MONTH)\
                                                   & (not_f_allocated['Week number'] == WEEK_NUMBER)
                                                   , ['Acc']])['Acc'].unique()

idle_acc_cond = (idle.loc[(idle['Month'] == MONTH)\
                           & (idle['Week number'] == WEEK_NUMBER)
                           , ['Acc']])['Acc'].unique()

```
