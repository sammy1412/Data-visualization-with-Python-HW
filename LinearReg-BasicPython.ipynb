```python
# plot data
import matplotlib.pyplot as plt
import random

areas  = [6.7, 4.6, 3.5, 5.1, 4.4, 4.0, 6.6, 3.6, 6.3, 5.8, 2.8, 5.2]
prices = [7.4, 5.9, 4.6, 6.0, 5.5, 5.7, 7.9, 5.5, 7.5, 6.6, 4.1, 6.3]
N = len(areas)

plt.scatter(areas, prices)
plt.xlabel('Area (x 100$m^2$)')
plt.ylabel('Price ($1000)')
plt.show()
```

####  Implementation

```python
# shuffle data
def shuffle_data(X, y):
    zip_data = list(zip(X, y))
    random.shuffle(zip_data)
    X_shuffled, y_shuffled = zip(*zip_data)

    return X_shuffled, y_shuffled

# forward
def predict(x, w, b):
    return x*w + b

# compute gradient
def gradient(y_hat, y, x):
    dw = 2*x*(y_hat-y)
    db = 2*(y_hat-y)
    
    return (dw, db)

# update weights
def update_weight(w, b, lr, dw, db):
    w_new = w - lr*dw
    b_new = b - lr*db
    
    return (w_new, b_new)
```

```python
# data preparation
print(f'areas: {areas}')
print(f'prices: {prices}')
print(f'data_size: {N}')
```

```python
# init weights
b = 1
w = 0.5
lr = 0.01

# parameter
epoch_max = 3
losses = [] # monitoring
parameters = [] # monitoring

for epoch in range(epoch_max):
    # shuffle data
    areas_sh, prices_sh = shuffle_data(areas, prices)
    
    # for an epoch    
    for i in range(N):
        
        # get a sample
        x = areas_sh[i]
        y = prices_sh[i]      
        
        # predict y_hat
        y_hat = predict(x, w, b)

        # compute loss
        loss = (y_hat-y)**2
        
        # for debug
        losses.append(loss)

        # compute gradient
        (dw, db) = gradient(y_hat, y, x)

        # update weights
        parameters.append((w, b))
        (w, b) = update_weight(w, b, lr, dw, db)
    lr *= .67
```

```python
learn_n = 0
for _ in parameters:
    print('Lần học thứ {}, Loss = {}'.format(learn_n + 1, losses[learn_n]))
    w = _[0]
    b = _[1]
    prices_pred = [x*w + b for x in areas]
    plt.plot(areas, prices_pred, 'r')
    plt.scatter(areas, prices)
    plt.show()
    learn_n += 1
```
